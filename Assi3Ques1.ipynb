```python
import numpy as np
```

```python
grid_size = 4
number_of_actions = 4
non_terminal = list(range(2,grid_size*grid_size))
terminal = [1,grid_size*grid_size]
reward = -1 #for all states

#since equiprobable so probability of going in 1 direction is 0.25
action_prob = 0.25

actions = ['UP','DOWN','LEFT','RIGHT']
```

```python
# since  we can move up down left right so lets define a function which will give us next state
```

```python
def next_state(current_state , action):
  row,col = current_state // grid_size , current_state % grid_size

  if action =='UP':
    row = max(0,row-1)
  elif action == 'DOWN':
    row = min(row+1,grid_size-1)
  elif action == 'LEFT':
    col = max(col-1,0)
  elif action == 'RIGHT':
    col = min(col+1,grid_size-1)
  return row*grid_size + col+1  # a state in form of row and column

```

```python
def policy_evaluation():
  #since it is random policy so we initialize all values to be zero
  #initialization
  V = np.zeros(grid_size*grid_size+1)
  V[terminal] = 0   #terminal states have 0 values

  #policy
  iteration = 0
  while True:
    delta = 0

    for state in range(1,grid_size*grid_size+1):  #visit every state
      if state not in terminal:
        v = V[state]
        new_v = 0;
        for action in actions:
          next_state_after_action = next_state(state,action)
          new_v += action_prob*(reward+V[next_state_after_action])
        V[state] = new_v
        delta = max(delta,abs(v-V[state]))
    iteration += 1
    print(f"Iteration {iteration}- STATE_VALUES:")
    print(V[1:].reshape((grid_size*grid_size)))
    if delta < 1e-6:
        break

  return V[1:].reshape((grid_size*grid_size))




# print("state values of grid are ")
# print(state_value)
```

```python
state_value = policy_evaluation()
```
